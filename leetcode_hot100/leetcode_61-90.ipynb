### 96. Unique Binary Search Trees(Median)

Given an integer n, return the number of structurally unique BST's (binary search trees) which has exactly n nodes of unique values from 1 to n.

<img src="https://assets.leetcode.com/uploads/2021/01/18/uniquebstn3.jpg" width="500px">

Example 1:

Input: n = 3
Output: 5

Example 2:

Input: n = 1
Output: 1



**Hint:**

    - DP Problem
    - Divide Tree into `left subtree` and `right subtree`, f(N) = sum(prod(f(N-1-i)*f(i)))

````python
class Solution(object):
    def numTrees(self, n):
        """
        :type n: int
        :rtype: int
        """

        rst = [1, 1, 2]

        if n <= 2:
            return rst[n]
        else:
            iter_ = 3
            while iter_ <= n:
                rst.append(sum([rst[i]*rst[-(1+i)] for i in range(len(rst))]))
                iter_ += 1
            
        return rst[-1]
````

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """

        queue, queue_next, values = [root], [], []

        while queue:
            # print(queue, queue_next)
            for item in queue:
                values.append(item.val)
                if item.val != "null":
                    queue_next.append(item.left if item.left else TreeNode("null"))
                    queue_next.append(item.right if item.right else TreeNode("null"))

            if values != values[::-1]:
                return False

            queue = queue_next
            # print(queue, queue_next)
            queue_next = []
            values = []

        return True
````

### 104. Maximum depth in Binary Tree (Simple)

Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

 

Example 1:

<img src="https://assets.leetcode.com/uploads/2020/11/26/tmp-tree.jpg" width=300>

Input: root = [3,9,20,null,null,15,7]
Output: 3

Example 2:
Input: root = [1,null,2]
Output: 2
 

Constraints:

The number of nodes in the tree is in the range [0, 104].
-100 <= Node.val <= 100

**Hint:**

```
Tried dfs and wfs two methods, this one is simple
```

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """

        def find_depth(cur, node):
            
            if node is None or node.val == "null":
                return cur
            else:
                cur += 1
                return max(find_depth(cur, node.left), find_depth(cur, node.right))

        return find_depth(0, root)


class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """

        if not root:
            return 0

        queue, queue_next = [root], []
        depth = 0

        while queue:
            for i in queue:
                if i.val != "null":
                    if i.left:
                        queue_next.append(i.left)
                    if i.right:
                        queue_next.append(i.right)
            
            depth += 1
            queue = queue_next
            queue_next = []

        return depth

````

### 105. Construct Binary Tree from Preorder and Inorder Traversal(Median)

Given two integer arrays preorder and inorder where preorder is the preorder traversal of a binary tree and inorder is the inorder traversal of the same tree, construct and return the binary tree.

Example 1:

<img src="https://assets.leetcode.com/uploads/2021/02/19/tree.jpg" width=300 height=260>

Input: preorder = [3,9,20,15,7], inorder = [9,3,15,20,7]
Output: [3,9,20,null,null,15,7]

Example 2:

Input: preorder = [-1], inorder = [-1]
Output: [-1]

**Hint**
```
preorder's first point is always root, split inorder/preorder into left and right, recursive
```

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: TreeNode
        """

        def build(preorder, inorder, tree=None):

            # print(tree)

            if not preorder:
                return tree

            node_cur = preorder[0]
            tree = TreeNode(node_cur)
            # tree.val = node_cur

            p_pre, p_in = 1, 0
            preorder_left, preorder_right = [], []
            inorder_left, inorder_right = [], []
            
            while inorder[p_in] != node_cur:
                inorder_left.append(inorder[p_in])
                preorder_left.append(preorder[p_pre])

                p_in += 1
                p_pre += 1
            
            preorder_right = preorder[p_pre:]
            inorder_right = inorder[p_pre:]

            tree.left = build(preorder_left, inorder_left)
            tree.right = build(preorder_right, inorder_right)

            return tree

        tree = build(preorder, inorder)

        return tree
````

### 114. Flatten Binary Tree to Linked List(Median)

Given the root of a binary tree, flatten the tree into a "linked list":

**Please do that in O(1) Space complexity**

The "linked list" should use the same TreeNode class where the right child pointer points to the next node in the list and the left child pointer is always null.

The "linked list" should be in the same order as a pre-order traversal of the binary tree.
 

Example 1:

<img src="https://assets.leetcode.com/uploads/2021/01/14/flaten.jpg" width=300>

Input: root = [1,2,5,3,4,null,6]

Output: [1,null,2,null,3,null,4,null,5,null,6]

Example 2:

Input: root = []

Output: []

Example 3:

Input: root = [0]

Output: [0]
 

Constraints:

The number of nodes in the tree is in the range [0, 2000].
-100 <= Node.val <= 100


**Hint:**

  - Find left branch's deepest right node, point to right branch
  - Copy left branch to right and set left branch None
  - Iterate

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def flatten(self, root):
        """
        :type root: TreeNode
        :rtype: None Do not return anything, modify root in-place instead.
        """

        def move_left_deep(node):
            
            while node.left or node.right:

                if node.right is not None:
                    node = node.right
                else:
                    node = node.left

            return node
        
        if root is None:
            return

        current = root

        while current.left or current.right:

            if current.left is None:
                pass
            else:
                left_end = move_left_deep(current.left)
                left_end.left = None
                left_end.right = current.right
                
                current.right = current.left
                current.left = None

            current = current.right
````

### 121. Best Time to Buy and Sell Stock(Simple)

You are given an array prices where prices[i] is the price of a given stock on the ith day.

You want to maximize your profit by choosing a single day to buy one stock and choosing a different day in the future to sell that stock.

Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.

**Fix in O(n)**

Example 1:

Input: prices = [7,1,5,3,6,4]
Output: 5
Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.
Note that buying on day 2 and selling on day 1 is not allowed because you must buy before you sell.
Example 2:

Input: prices = [7,6,4,3,1]
Output: 0
Explanation: In this case, no transactions are done and the max profit = 0.
 

Constraints:

1 <= prices.length <= 105
0 <= prices[i] <= 104


````python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """

        profit = 0
        n = len(prices)
        start, end = 0, 1
        cur_min = prices[start]


        while end < n:
            
            if prices[end] - prices[start] > profit:
                profit = prices[end] - prices[start]
            
            if prices[end] < cur_min:
                start = end
                cur_min = prices[end]

            end += 1

        return profit
````
