### 96. Unique Binary Search Trees(Median)

Given an integer n, return the number of structurally unique BST's (binary search trees) which has exactly n nodes of unique values from 1 to n.

<img src="https://assets.leetcode.com/uploads/2021/01/18/uniquebstn3.jpg" width="500px">

Example 1:

Input: n = 3

Example 2:

Input: n = 1
Output: 1



**Hint:**

    - DP Problem
    - Divide Tree into `left subtree` and `right subtree`, f(N) = sum(prod(f(N-1-i)*f(i)))

````python
class Solution(object):
    def numTrees(self, n):
        """
        :type n: int
        :rtype: int
        """

        rst = [1, 1, 2]

        if n <= 2:
            return rst[n]
        else:
            iter_ = 3
            while iter_ <= n:
                rst.append(sum([rst[i]*rst[-(1+i)] for i in range(len(rst))]))
                iter_ += 1
            
        return rst[-1]
````

<img src="https://assets.leetcode.com/uploads/2020/11/26/tmp-tree.jpg" width=300>

Input: root = [3,9,20,null,null,15,7]

Constraints:

-100 <= Node.val <= 100

**Hint:**

```
Tried dfs and wfs two methods, this one is simple
```

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """

        def find_depth(cur, node):
            
            if node is None or node.val == "null":
                return cur
            else:
                cur += 1
                return max(find_depth(cur, node.left), find_depth(cur, node.right))

        return find_depth(0, root)


class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """

        if not root:
            return 0

        queue, queue_next = [root], []
        depth = 0

        while queue:
            for i in queue:
                if i.val != "null":
                    if i.left:
                        queue_next.append(i.left)
                    if i.right:
                        queue_next.append(i.right)
            
            depth += 1
            queue = queue_next
            queue_next = []

        return depth

````

<img src="https://assets.leetcode.com/uploads/2021/02/19/tree.jpg" width=300 height=260>

Example 2:

Input: preorder = [-1], inorder = [-1]
Output: [-1]

**Hint**
```
preorder's first point is always root, split inorder/preorder into left and right, recursive
```

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: TreeNode
        """

        def build(preorder, inorder, tree=None):

            # print(tree)

            if not preorder:
                return tree

            node_cur = preorder[0]
            tree = TreeNode(node_cur)
            # tree.val = node_cur

            p_pre, p_in = 1, 0
            preorder_left, preorder_right = [], []
            inorder_left, inorder_right = [], []
            
            while inorder[p_in] != node_cur:
                inorder_left.append(inorder[p_in])
                preorder_left.append(preorder[p_pre])

                p_in += 1
                p_pre += 1
            
            preorder_right = preorder[p_pre:]
            inorder_right = inorder[p_pre:]

            tree.left = build(preorder_left, inorder_left)
            tree.right = build(preorder_right, inorder_right)

            return tree

        tree = build(preorder, inorder)

        return tree
````

<img src="https://assets.leetcode.com/uploads/2021/01/14/flaten.jpg" width=300>

Constraints:

-100 <= Node.val <= 100


**Hint:**

  - Copy left branch to right and set left branch None
  - Iterate

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def flatten(self, root):
        """
        :type root: TreeNode
        :rtype: None Do not return anything, modify root in-place instead.
        """

        def move_left_deep(node):
            
            while node.left or node.right:

                if node.right is not None:
                    node = node.right
                else:
                    node = node.left

            return node
        
        if root is None:
            return

        current = root

        while current.left or current.right:

            if current.left is None:
                pass
            else:
                left_end = move_left_deep(current.left)
                left_end.left = None
                left_end.right = current.right
                
                current.right = current.left
                current.left = None

            current = current.right
````

### 121. Best Time to Buy and Sell Stock(Simple)

You are given an array prices where prices[i] is the price of a given stock on the ith day.

You want to maximize your profit by choosing a single day to buy one stock and choosing a different day in the future to sell that stock.

Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.

**Fix in O(n)**

Example 1:

Input: prices = [7,1,5,3,6,4]
Output: 5
Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.
Note that buying on day 2 and selling on day 1 is not allowed because you must buy before you sell.
Example 2:

Input: prices = [7,6,4,3,1]
Output: 0
Explanation: In this case, no transactions are done and the max profit = 0.
 

Constraints:

1 <= prices.length <= 105
0 <= prices[i] <= 104


````python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """

        profit = 0
        n = len(prices)
        start, end = 0, 1
        cur_min = prices[start]


        while end < n:
            
            if prices[end] - prices[start] > profit:
                profit = prices[end] - prices[start]
            
            if prices[end] < cur_min:
                start = end
                cur_min = prices[end]

            end += 1

        return profit
````

### 124. Binary Tree Maximum Path Sum(Hard)

A path in a binary tree is a sequence of nodes where each pair of adjacent nodes in the sequence has an edge connecting them. A node can only appear in the sequence at most once. Note that the path does not need to pass through the root.

The path sum of a path is the sum of the node's values in the path.

Given the root of a binary tree, return the maximum path sum of any non-empty path.

Explanation: The optimal path is 15 -> 20 -> 7 with a path sum of 15 + 20 + 7 = 42.
 

Constraints:

The number of nodes in the tree is in the range [1, 3 * 104].
-1000 <= Node.val <= 1000

**Hint**

  - Calculate maximum(left+root, right+root, root) as new_node_val, put [left, right, left+right+root] into cache
  - Recursive

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxPathSum(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """

        def subTreeSum(tree, sums=[]):

            if not (tree.left or tree.right):
                sums.append(tree.val)
                return tree, sums


            if tree.left is not None:
                tree.left, sums = subTreeSum(tree.left, sums)
            if tree.right is not None:
                tree.right, sums = subTreeSum(tree.right, sums)


            sums.append(sum([
                    tree.left.val if tree.left else 0, 
                    tree.right.val if tree.right else 0,
                    tree.val
                ]))
            
            sums.append(tree.left.val if tree.left else -1001)
            sums.append(tree.right.val if tree.right else -1001)

            tree.val += max(
                    tree.left.val if tree.left else 0, 
                    tree.right.val if tree.right else 0,
                    0
                ) 


            tree.left = None
            tree.right = None


            return tree, sums


        root, sums = subTreeSum(root, sums=[])
        return max(root.val, max(sums))


````

### 128. Longest Consecutive Sequence(Median)

Given an unsorted array of integers nums, return the length of the longest consecutive elements sequence.

You must write an algorithm that runs in O(n) time.

 

Example 1:

Input: nums = [100,4,200,1,3,2]

Output: 4

Explanation: The longest consecutive elements sequence is [1, 2, 3, 4]. Therefore its length is 4.

Example 2:

Input: nums = [0,3,7,2,5,8,4,6,0,1]

Output: 9


````python
class Solution:
    def longestConsecutive(self, nums):
        longest_streak = 0
        num_set = set(nums)

        for num in num_set:
            if num - 1 not in num_set:
                current_num = num
                current_streak = 1

                while current_num + 1 in num_set:
                    current_num += 1
                    current_streak += 1

                longest_streak = max(longest_streak, current_streak)

        return longest_streak
````

### 136 Single Number(Easy)

Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.

Example 1:

Input: nums = [2,2,1]

Output: 1

Example 2:

Input: nums = [4,1,2,1,2]

Output: 4

Example 3:

Input: nums = [1]

Output: 1


**Hint:**
 - Although this is easy, it is actually interesting
 - Use `XOR (^ in python)` , it means compare XOR of every digits in binary system
 - Contribution of `XOR`
   - a ^ b <=> b ^ a
   - (a ^ b) ^ c <=> a ^ (b ^ c)
   - 0 ^ n => n
   - n ^ n => 0
 - a ^ b ^ c ^ a ^ b <=> a ^ a ^ b ^ b ^ c <=> 0 ^ 0 ^ c <=> c


````python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        rst = 0

        for i in nums:
            rst ^= i

        return rst
````

Given the head of a linked list, return the node where the cycle begins. If there is no cycle, return null.

There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. Internally, pos is used to denote the index of the node that tail's next pointer is connected to (0-indexed). It is -1 if there is no cycle. Note that pos is not passed as a parameter.

**Hint:**

  - fast and slow point
    - if cycle, fast will meet slow
    - if cycle, fast will meet slow in one lap.  c/(2-1) < b + c 
    - 2*(a + b) = a + n * (b + c) + b, so a = n * (b + c) - b = (n - 1) * (b + c) + c,

````python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def detectCycle(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """

        if head is None or head.next is None:
            return 

        start = head
        fast, slow = head, head

        rst = True
        while rst:
            fast = fast.next.next
            slow = slow.next

            if fast is None or fast.next is None:
                rst = False
                break
            elif slow is fast:
                break

        loop_set = set([])

        while rst:
            if fast not in loop_set:
                loop_set.add(fast)
                fast = fast.next
            else:
                while start is not None:
                    if start in loop_set:
                        return start
                    else:
                        start = start.next
        
        return 
        


            
````
